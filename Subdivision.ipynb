```python
"""
We want to subdivide (ie. make really smooth) a 3d mesh based on Catmull-Clark given that the mesh is closed and has only quad faces.
"""
```

```python
import numpy as np
```

```python
# parse mesh file
f=open('mod.off','r')
lines=f.readlines()
(nvert,nface,_) = lines[1].split(' ')
verts=[]
for i in range(2,2+int(nvert)) :
    (x,y,z)=lines[i].split()
    verts.append( (float(x.strip()),float(y.strip()),float(z.strip())) )
faces=[]
for i in range(2+int(nvert),2+int(nvert)+int(nface)) :
    (n,a,b,c,d)=lines[i].split()
    faces.append( (int(a.strip()),int(b.strip()),int(c.strip()),int(d.strip())) )
```

```python
# create all face pts for subdivision
facepts={}
for f in faces :
    nx=(verts[f[0]][0]+verts[f[1]][0]+verts[f[2]][0]+verts[f[3]][0])/4
    ny=(verts[f[0]][1]+verts[f[1]][1]+verts[f[2]][1]+verts[f[3]][1])/4
    nz=(verts[f[0]][2]+verts[f[1]][2]+verts[f[2]][2]+verts[f[3]][2])/4
    facepts[f]=(nx,ny,nz)
```

```python
#make the subdivided point for each edge
def makeEdgePt(v1,v2) :
    key=(min(v1,v2),max(v1,v2))
    if key in edgepts : #already looked at this edge
        return
    #get two faces that contain both of these vertices (ie. this edge)
    these_face_pts=[]
    for f in faces :
        if v1 in f and v2 in f :
            these_face_pts.append(f)
    nx = (verts[v1][0]+verts[v2][0]+facepts[these_face_pts[0]][0]+facepts[these_face_pts[1]][0])/4
    ny = (verts[v1][1]+verts[v2][1]+facepts[these_face_pts[0]][1]+facepts[these_face_pts[1]][1])/4
    nz = (verts[v1][2]+verts[v2][2]+facepts[these_face_pts[0]][2]+facepts[these_face_pts[1]][2])/4
    edgepts[key]=(nx,ny,nz)
    
#create 4 edge points for each face if we haven't already looked at that edge in another face
edgepts={}
for f in faces :
    (a,b,c,d)=f
    makeEdgePt(a,b)
    makeEdgePt(b,c)
    makeEdgePt(c,d)
    makeEdgePt(d,a)
```

```python
#make an edge representation for later computation
edges=[]
for f in faces :
    (a,b,c,d)=f
    if (a,b) not in edges :
        edges.append((a,b))
    if (b,a) not in edges :
        edges.append((b,a))
    if (b,c) not in edges :
        edges.append((b,c))
    if (c,b) not in edges :
        edges.append((c,b))
    if (c,d) not in edges :
        edges.append((c,d))
    if (d,c) not in edges :
        edges.append((d,c))
    if (d,a) not in edges :
        edges.append((d,a))
    if (a,d) not in edges :
        edges.append((a,d))
```

```python
# make the new vertex points for the subdivision
new_verts=[]
for i in range(int(nvert)) :
    #get every edge and every face that has that vertex
    centers=[]
    for f in faces :
        (a,b,c,d)=f
        if i==a or i==b or i==c or i==d:
            centers.append(facepts[f])
            
    edges_included=[]
    for (a,b) in edges :
        if a==i or b==i :
            edges_included.append((min(a,b),max(a,b)))
    edges_included=list(set(edges_included))

    #calculate summations for each type of point (face, edge, vertex)
    nv=[]
    for axis in range(3) :
        n=len(centers)
        #print n
        Q=0
        for j in range(n) :
            Q += centers[j][axis]
        Q/=n
        R=0
        for j in range(n) :
            R += (verts[edges_included[j][0]][axis]+verts[edges_included[j][1]][axis])/2
        R/=n
        P=verts[i][axis]
        # calculate the new point based on the equation that satisfies cubic smoothing properties
        nv.append( ( Q+R*2+P*(n-3) )/n )
    new_verts.append(tuple(nv))
```

```python
# form new groups of 4 points as faces

# get id of vertex and otherwise make a new id
def getid(v) :
    if v in vert_pos2id.keys() :
        return vert_pos2id[v]
    id=len(vert_pos2id)
    vert_pos2id[v]=id
    return id
    
vert_pos2id={} #map of vertices to ids because when we print faces we represent them as 4-tuples of integer ids
newfaces=[]
for f in faces :
    (a,b,c,d)=f
    facept=facepts[f]
    e1=edgepts[(min(a,b),max(a,b))]
    e2=edgepts[(min(b,c),max(b,c))]
    e3=edgepts[(min(c,d),max(c,d))]
    e4=edgepts[(min(d,a),max(d,a))]
    v1=new_verts[a]
    v2=new_verts[b]
    v3=new_verts[c]
    v4=new_verts[d]
    facept_id=getid(facept)
    
    #get point ids
    e1_id=getid(e1)
    e2_id=getid(e2)
    e3_id=getid(e3)
    e4_id=getid(e4)
    v1_id=getid(v1)
    v2_id=getid(v2)
    v3_id=getid(v3)
    v4_id=getid(v4)
    
    #make the representation of new faces
    newfaces.append((v1_id,e1_id,facept_id,e4_id))
    newfaces.append((v2_id,e2_id,facept_id,e1_id))
    newfaces.append((v3_id,e3_id,facept_id,e2_id))
    newfaces.append((v4_id,e4_id,facept_id,e3_id))
    
# reverse the mapping of vert_pos2id so we have id: point
verts2=dict (zip(vert_pos2id.values(),vert_pos2id.keys())).values()
faces2=newfaces
```
